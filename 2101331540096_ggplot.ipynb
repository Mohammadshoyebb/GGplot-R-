```r
library(ggplot2)
```

#Creating a Dataframe

```r
library("dplyr")

library("tidyverse")


emp_id=c(96,120,63,23,83,45,85,47,98,74)
emp_name=c("Satyajeet","Ujjawal","Arman","Arnav","Ramit","Ram","Sita","Lakshman","Hanuman","Krishna")
emp_age=c(21,22,23,31,45)
emp_sal=c(50000,60000,70000,45000,80000,63000,41000,85000,65478,12000)
emp_desig=c("AP","AP","PROF","DIRECTOR","DEAN","AP","HOD","Dy.Hod","President","Registrar")
city=c("Noida","Bombay","Pune","Goa","Ranchi","Dhanbad","Noida","Chennai","Gurugram","Goa")
dept=c("DS","AI","CSBS","DS","AIML","CYBER","MCA","MBA","CSE","IT")
pin=c(826001,828130,201310,201301,854678,826001,828130,456987,201310,200456)
dis=c(12,56,78,95,89,24,57,68,89,5)



```

```r
emp=data.frame(emp_id,emp_name,emp_age,emp_sal,emp_desig,city,dept,pin,dis)
emp
```

```r
Hra <- c(201,202,203,204,205,206,207,208,209,210)
Da <- c(350,380,280,340,560,340,560,230,760,870)
Ta <- c(2011,2012,2013,2014,2015,2016,2017,2018,2019,2110)
Basic <- c(23000, 34000,40000, 34500,12000,24000,78000,45000,34000,60000)
Tax <- c(3500,3800,2800,3400,5600,3400,5600,2300,7600,8700)
emp_sal = data.frame(emp_id,Hra,Da,Ta,Basic,Tax)
emp_sal

```

#Merging Two Dataframes

```r

emp_detail = merge.data.frame(emp,emp_sal)
emp_detail
```

Basic Plotting Syntax

```r
ggplot(data=emp_detail,aes(x=emp_id,y=emp_sal))
```

**Creating plot using geom_point() Function.**

```r
ggplot(data=emp_detail,aes(x=emp_id,y=emp_sal))+geom_point(colour="blue",size=10)
```

Creating line plot using geom_line() Function.

```r
ggplot(data=emp_detail,aes(x=emp_age,y=emp_sal))+geom_line(colour="red")
```

**Creating bar plot using geom_col() function**

```r
ggplot(data=emp_detail,aes(x=emp_name,y=emp_age))+geom_col()
```

```r
ggplot(data=emp_detail,aes(x=city,y=emp_sal))+geom_col(color="green")
```

**Creating bar plot using geom_bar() function**

```r
ggplot(data=emp_detail,aes(x=city,y=emp_sal))+geom_bar(color="blue",stat="summary",fun="mean")
```

```r
ggplot(data=emp_detail,aes(x=city,y=emp_sal))+geom_bar(color="red",stat="summary",fun="median")
```

```r
#Stacked and grouped bar charts
ggplot(data=emp_detail,aes(x=city,y=emp_sal))+geom_col(color="red",stat="summary",fun="median",position="dodge")
```

**Plotting Histograms**

```r
ggplot(emp_detail,aes(x=emp_id))+geom_histogram(binwidth = 0.6, fill="yellow", colour="black")
```

```r
#plotting Distributions

ggplot(emp_detail,aes(x=emp_sal,fill=city))+geom_density(colour="red",alpha=0.5)
```

```r
#Violin plots
```

```r
ggplot(emp_detail,aes(x=emp_age, y=emp_sal))+geom_violin(colour="black",fill="red")
```

```r
#Boxplot
ggplot(emp_detail,aes(x=emp_age, y=emp_sal))+geom_boxplot(colour="black",fill="skyblue")
```

```r
#stripcharts

ggplot(emp_detail,aes(x=emp_age, y=Tax))+geom_jitter(height=0,width=0.3)
```
